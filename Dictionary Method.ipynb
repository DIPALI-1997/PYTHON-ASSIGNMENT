```python
import logging
logging.basicConfig(filename='personalinfo.log', level=logging.INFO)

class PersonInfo:
    def __init__(self):
        self.person_dict = {}

    def ask_for_input(self):
        try:
            self.person_dict['name'] = input("Enter the person's name: ")
            self.person_dict['age'] = int(input("Enter the person's age: "))
            self.person_dict['city'] = input("Enter the person's city: ")
        except ValueError:
            logging.error("Invalid input for age. Please enter a number.")
            print("Invalid age input. Please enter a number.")
            return False
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")
            print("An error occurred. Please try again.")
            return False
        return True

    def create_person_dict(self):
        if self.ask_for_input():
            print("Person information:", self.person_dict)

# Usage
person_info = PersonInfo()
person_info.create_person_dict()
```

```python
import logging
logging.basicConfig(filename='citypopulation.log', level=logging.INFO)

class CityPopulations:
    def __init__(self):
        self.cities = {}

    def create_and_update_cities(self):
        try:
            while True:
                action = input("Enter 'add' to add a city, 'update' to update population, or 'done' to finish: ")
                if action.lower() == "done":
                    break

                if action.lower() == "add":
                    self.add_city()
                elif action.lower() == "update":
                    self.update_population()
                else:
                    print("Invalid action. Please choose 'add', 'update', or 'done'.")
        except Exception as e:
            logging.error("An error occurred:", e)
            print("An error occurred. Please try again.")
        else:
            self.print_cities()

    def add_city(self):
        try:
            city_name = input("Enter the city name: ")
            population = int(input("Enter the population: "))
            self.cities[city_name] = population
            print(f"City {city_name} added with population {population}")
        except ValueError:
            print("Invalid input for population. Please enter a number.")

    def update_population(self):
        city_name = input("Enter the city name to update: ")
        if city_name in self.cities:
            try:
                new_population = int(input("Enter the new population: "))
                self.cities[city_name] = new_population
                print(f"Population of {city_name} updated to {new_population}")
            except ValueError:
                print("Invalid input for population. Please enter a number.")
        else:
            print(f"City {city_name} not found in the dictionary.")

    def print_cities(self):
        print("Final City Populations:")
        for city, population in self.cities.items():
            print(f"{city}: {population}")

# Usage
city_populations = CityPopulations()
city_populations.create_and_update_cities()
```

```python
import logging
logging.basicConfig(filename='MonthAndDays.log', level=logging.INFO)

class MonthsAndDays:
    def __init__(self):
        self.months_dict = {
            "January": 31,
            "February": 28,  # Assuming a non-leap year
            "March": 31,
            "April": 30,
            "May": 31,
            "June": 30,
            "July": 31,
            "August": 31,
            "September": 30,
            "October": 31,
            "November": 30,
            "December": 31
        }

    def interact_with_user(self):
        try:
            while True:
                action = input("Enter 'month' to get days in a month, 'all' to list all months, or 'done' to finish: ")
                if action.lower() == 'done':
                    break
                elif action.lower() == 'month':
                    self.print_days_in_month()
                elif action.lower() == 'all':
                    self.print_all_months_and_days()
                else:
                    print("Invalid action. Please choose 'month', 'all', or 'done'.")
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")
            print("An error occurred. Please try again.")

    def print_days_in_month(self):
        month = input("Enter the month: ")
        month = month.capitalize()
        if month in self.months_dict:
            print(f"{month} has {self.months_dict[month]} days.")
        else:
            print(f"{month} is not a valid month.")

    def print_all_months_and_days(self):
        print("Months and Days:")
        for month, days in self.months_dict.items():
            print(f"{month}: {days} days")

# Usage
months_and_days = MonthsAndDays()
months_and_days.interact_with_user()
```

```python
import logging
logging.basicConfig(filename='namelenght.log', level=logging.INFO)

class NameLengths:
    def __init__(self):
        pass

    def create_name_length_dict(self):
        try:
            names_str = input("Enter names separated by commas: ")
            # Split the string into names and strip spaces from each name
            names = [name.strip() for name in names_str.split(",")]

            # Create a dictionary with names and their lengths
            name_lengths = {name: len(name) for name in names}

            # Print the final dictionary
            print("Name lengths dictionary:", name_lengths)
        except Exception as e:
            logging.error("An error occurred while creating the dictionary:", e)
            print("An error occurred. Please try again.")

# Usage
name_lengths = NameLengths()
name_lengths.create_name_length_dict()
```

```python
import logging
logging.basicConfig(filename='Itemprices.log', level=logging.INFO)

class ItemPrices:
    def __init__(self):
        self.items_dict = {}

    def create_and_print_items(self):
        try:
            while True:
                action = input("Enter 'add' to add an item, 'print' to print items, or 'done' to finish: ")
                if action.lower() == "done":
                    break

                if action.lower() == "add":
                    self.add_item()
                elif action.lower() == "print":
                    self.print_items()
                else:
                    print("Invalid action. Please choose 'add', 'print', or 'done'.")
        except Exception as e:
            logging.error(f"An error occurred: {e}")
            print("An error occurred. Please try again.")

    def add_item(self):
        try:
            name = input("Enter the item name: ")
            price = float(input("Enter the price: "))
            self.items_dict[name] = price
            print(f"Item {name} added with price ${price:.2f}")
        except ValueError:
            print("Invalid input for price. Please enter a valid number.")

    def print_items(self):
        if self.items_dict:
            print("Items and prices:")
            for item, price in self.items_dict.items():
                print(f"- {item}: ${price:.2f}")
        else:
            print("No items in the list.")

# Usage
item_prices = ItemPrices()
item_prices.create_and_print_items()
```

```python
import logging
logging.basicConfig(filename='keychecker.log', level=logging.INFO)

class KeyChecker:
    def __init__(self):
        self.my_dict = {}  # Initialize an empty dictionary

    def add_key_value(self):
        try:
            key = input("Enter the key to add: ")
            value = input("Enter the value for the key: ")
            self.my_dict[key] = value
            print(f"Key '{key}' with value '{value}' has been added to the dictionary.")
        except Exception as e:
            logging.error("An error occurred while adding the key-value pair:", e)
            print("An error occurred. Please try again.")

    def check_key(self):
        try:
            key_to_check = input("Enter the key to check for: ")

            if key_to_check in self.my_dict:
                print(f"Key '{key_to_check}' is present in the dictionary.")
            else:
                print(f"Key '{key_to_check}' is not present in the dictionary.")
        except Exception as e:
            logging.error("An error occurred while checking for the key:", e)
            print("An error occurred. Please try again.")

    def interact_with_user(self):
        try:
            while True:
                action = input("Enter 'add' to add a key-value pair, 'check' to check for a key, or 'done' to finish: ")
                if action.lower() == "done":
                    break

                if action.lower() == "add":
                    self.add_key_value()
                elif action.lower() == "check":
                    self.check_key()
                else:
                    print("Invalid action. Please choose 'add', 'check', or 'done'.")
        except Exception as e:
            logging.error("An error occurred:", e)
            print("An error occurred. Please try again.")

# Usage
key_checker = KeyChecker()
key_checker.interact_with_user()
```

```python
import logging
logging.basicConfig(filename='WordlengthSorter.log', level=logging.INFO)

class WordLengthSorter:
    def __init__(self):
        self.word_lengths = {}

    def create_and_sort_dict(self):
        try:
            words_str = input("Enter words separated by commas: ")
            # Split the string into words and strip spaces from each word
            words = [word.strip() for word in words_str.split(",")]

            # Create a dictionary with words and their lengths
            self.word_lengths = {word: len(word) for word in words}

            # Sort the dictionary by word length
            sorted_word_lengths = sorted(self.word_lengths.items(), key=lambda item: item[1])

            # Print the sorted words by length
            print("Sorted words by length:")
            for word, length in sorted_word_lengths:
                print(f"- {word} ({length})")
        except Exception as e:
            logging.error(f"An error occurred: {e}")
            print("An error occurred. Please try again.")

# Usage
word_length_sorter = WordLengthSorter()
word_length_sorter.create_and_sort_dict()
```

```python
import logging
logging.basicConfig(filename='StudentScores.log', level=logging.INFO)
class StudentScores:
    def __init__(self):
        self.scores = {}

    def create_and_process_scores(self):
        try:
            while True:
                action = input("Enter 'add' to add a student score, 'print' to print scores and average, or 'done' to finish: ")
                if action.lower() == "done":
                    break

                if action.lower() == "add":
                    self.add_score()
                elif action.lower() == "print":
                    self.print_scores_and_average()
                else:
                    print("Invalid action. Please choose 'add', 'print', or 'done'.")
        except Exception as e:
            logging.error(f"An error occurred: {e}")
            print("An error occurred. Please try again.")

    def add_score(self):
        try:
            name = input("Enter student name: ")
            score = float(input("Enter score: "))
            self.scores[name] = score
            print(f"Score for {name} added.")
        except ValueError:
            print("Invalid input for score. Please enter a valid number.")

    def print_scores_and_average(self):
        if self.scores:
            total_score = sum(self.scores.values())
            class_average = total_score / len(self.scores)
            print("Student Scores:")
            for name, score in self.scores.items():
                print(f"- {name}: {score}")
            print(f"Class average: {class_average:.2f}")
        else:
            print("No scores added yet.")

# Usage
student_scores = StudentScores()
student_scores.create_and_process_scores()
```

```python
import logging

logging.basicConfig(filename='LibraryCatalog.log', level=logging.INFO)

class LibraryCatalog:
    def __init__(self): 
        self.catalog = {}

    def add_book(self, book_id, title, author, publication_year):
        self.catalog[book_id] = {
            "title": title,
            "author": author,
            "publication_year": publication_year
        }
        logging.info(f"Book added: {title}")

    @classmethod
    def get_book_info(cls, catalog, book_id):
        try:
            book = catalog[book_id]
            logging.info(f"Retrieved book info for {book_id}")
            return book
        except KeyError:
            logging.error(f"Book with ID {book_id} not found.")
            return None

# Example usage
library = LibraryCatalog()
library.add_book(1, "The Midnight Library", "Matt Haig", 2020)
library.add_book(2, "The Invisible Life of Addie LaRue", "V.E. Schwab", 2020)

book_info = LibraryCatalog.get_book_info(library.catalog, 1)
if book_info:
    print(f"Book Details: {book_info}")
else:
    print("Book not found.")
```
